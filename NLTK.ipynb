```python
import requests
from bs4 import BeautifulSoup as bs
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
import string
import nltk
import matplotlib.pyplot as plt
```

```python
# http://shakespeare.mit.edu/allswell/full.html
```

```python
# Specify the url
url= "http://shakespeare.mit.edu/allswell/full.html"

# Make a request to a web page and print the response text
try:
    r = requests.get(url)
    soup = bs(r.content, 'lxml')
except:
  pass
```

```python
text = soup.get_text()
text
```

### Normalizing Text


Converting the text to lowercase and remove newline characters '\n' from given text

```python
text = text.lower()
text = text.replace('\n','')
```

```python
text
```

```python
#tokenization
nltk.download('punkt')
nltk.download('wordnet')
nltk.download()
```

```python
#sentence level tokenization
sentences = sent_tokenize(text)
sentences
```

```python
len(sentences)
```

```python
sentences[0]
```

```python
#removing punctuations after sentence tokenization
for i,j in enumerate(sentences):
  print(i,j)
```

```python
sent_tokens=dict()
for i,j in enumerate(sentences):
    sent_tokens[i] = j.translate(str.maketrans('','',string.punctuation))
sent_tokens
```

```python
#avg sentence length
summ = 0
for i,j in enumerate(sentences):
  summ+=len(j.split(' '))
print(summ)
summ/len(sentences)
```

```python
words_len = []
for i,j in enumerate(sentences):
  words_len.append(len(j.split(' ')))

sum(words_len)/len(sentences)
```

word level tokenization

process of converting text into words




```python
words0=''
for i,j in enumerate(sentences):
    words0+=j.translate(str.maketrans('','',string.punctuation))
    # print(j.translate(str.maketrans('','',string.punctuation)))
words0
```

```python
words = word_tokenize(words0)
print(words)
```

Stemming
process of converting the words of a text tp its non-changing portions

```python
porter = nltk.PorterStemmer()
stemmed = [porter.stem(i) for i in words]
type(stemmed)
stemmed
```

Removing stopwords


```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
len(words)
```

```python
removed_stop_words=[]
for i in words:
  if i not in stop_words:
    removed_stop_words.append(i)
removed_stop_words
```

```python
len(removed_stop_words)
```

plot frequency distribution of top 25 words

```python
len(set(words))
```

```python
freq = nltk.FreqDist(removed_stop_words)
freq.plot(25)
```
