```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data collection and data analysis
Loading the data from it's csv file into pandas dataframe

```python
parkinson_data=pd.read_csv('/content/Parkinsson disease.csv')
print(parkinson_data)
```

```python
#no. of rows and columns in dataframe
parkinson_data.shape
```

```python
parkinson_data=pd.DataFrame(parkinson_data)
parkinson_data
```

Data pre-processing

```python
#seperating features and targets
X = parkinson_data.drop(columns=['name','status'], axis=1)
Y = parkinson_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting data into training n testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data standardization

```python
scaler =StandardScaler()
```

```python
scaler.fit(X_train)
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

Using SVM:Model training
SVM model

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train, Y_train)
```

Model evaluation
accuracy score


```python
# accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy score of training data : ', training_data_accuracy)
```

```python
#visualizing actual and predicted values
plt.bar(Y_train, X_train_prediction)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("actual vs predicted")
plt.show()
```

```python
#accuracy score on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy score of test data  using svm: ', testing_data_accuracy*100)
```


So by using svm algorithm, we achieve a maximum accuracy of 82%.

```python
#visualizing actual and predicted values
plt.bar(Y_test, X_test_prediction)
plt.xlabel("actual")
plt.ylabel("predicted")
plt.title("actual vs predicted")
plt.show()
```

XG Boost :

```python

#import XG Boost algorithms
from xgboost import XGBClassifier
```

```python
model= XGBClassifier(eval_metric='mlogloss')
model.fit(X_train, Y_train)
```

```python
#predicting the model
Y_pred= model.predict(X_test)
#accuracy score on testing data for xg boost
testing_data_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy score of testing data using XG Boost :',testing_data_accuracy*100)
```

```python
#using confusion matrix to predict no. of people affected by parkinson by  xgboost
from sklearn.metrics import confusion_matrix
pd.DataFrame(
    confusion_matrix(Y_test, Y_pred),
    columns= ['PREDICTED HEALTHY','Predicted parkinsons'],
    index = ['True healthy','True parkinsons']
)
```

```python
#visualizing using heatmap
sns.heatmap(confusion_matrix(Y_test , Y_pred),annot=True)
```

From the above heatmap we can see there are 30 predicted parkinsons as 1-->parkinsons positive

KNN model:

```python
#import knn classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
pca = PCA(n_components =2)
classifi = KNeighborsClassifier(n_neighbors =8,p=2,metric= 'minkowski')
classifi.fit(X_train, Y_train)
Y_pred = classifi.predict(X_test)
#accuracy score on testing data for knn
testing_data_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy score of testing data using KNN :',testing_data_accuracy*100)
```

Random Forest Classifier:

```python
#import random forest
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=30, max_depth=10, random_state= 1)
random_forest.fit(X_train, Y_train)
```

```python
Y_pred = random_forest.predict(X_test)
#accuracy score of testing data using random forest classifier
testing_data_accuracy = accuracy_score(Y_test, Y_pred)
print('Accuracy score of testing data using Random forest:', testing_data_accuracy*100)
```
