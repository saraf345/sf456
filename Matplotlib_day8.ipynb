```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[5,9,8,10,7,6,2]
```

to add marker use marker keyword,chng linestyle as well,chng colour of line also,

```python
plt.xlabel('days')
plt.ylabel('stocks')
plt.plot(x,y,marker='s',linestyle='--',color='blue',linewidth=3)
plt.title('sensex')
plt.show()
```

grids:

```python
plt.xlabel('Days')
plt.ylabel('Stocks')
plt.plot(x,y,marker='s',color='blue',linestyle='--',linewidth=5)
plt.title('sensex')
plt.grid(axis='y',linestyle=':',color='red',linewidth=2)
plt.show()
```

multi plots:

```python
import numpy as np
x=np.array([1,7,4,5])
y=np.array([7,8,9,6])
z=np.array([6,5,8,4])
plt.plot(x,label='x')
plt.plot(y,label='y')
plt.plot(z,label='z')
plt.grid(axis='x')
plt.show()
```

subplots[plots graohs seperately]

```python
#plot 1
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,3])
#plt.subplot(r,c,pos) syntax just keep changing pos
plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2
y=np.array([5,6,7,8,3])
z=np.array([4,6,3,7,9])
plt.subplot(1,2,2)
plt.plot(y,z)
plt.show()
```

print last dot of diff. clr

```python
import numpy as np
x=np.array([1,3,6,7,8])
y=x
color=['red']*(len(x)-1)
color.append('green')
plt.scatter(x,y,c=color,s=60,alpha=0.45)
plt.show()
```

bargraph

```python

x=np.array(['a','b','c'])
y=np.array([10,20,30])
plt.bar(x,y,color='brown',width=0.3)
plt.show
```

horizontal bars use barh and here* for width we use height here *

```python
x=np.array([6,7,8,9,10])
y=np.array([5,3,2,4,9])
plt.barh(x,y,color='yellow',height=0.3)
plt.show()
```
